## **Read total profit of all the months and show it using line plot.**

```python
%config InlineBackend.figure_format='retina'
plt.figure(figsize=(10, 6), dpi = 120)
plt.plot(data['month_number'], data['total_profit'])
plt.title('\nTotal Profit by Month\n', fontsize=25, fontweight = 'bold')
plt.xlabel('\nMonth\n', fontsize=15, fontweight = 'bold')
plt.ylabel(' Total Profit', labelpad = 65, rotation = 0, fontsize=15, fontweight = 'bold')
plt.grid(True)
plt.show()
```

# **Read all products sales data and show it using the multi-line plot.**

```python
%config InlineBackend.figure_format='retina'
plt.figure(figsize=(14, 8))
sns.set_palette("husl")

for product in ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']:
    plt.plot(data['month_number'], data[product], linewidth=3, marker='s', markersize=12, label=product)

plt.title('\nProduct Sales by Month\n', fontsize=28, fontweight='bold')
plt.xlabel('\nMonth\n', fontsize=20, fontweight = 'bold')
plt.ylabel(' Units Sold', labelpad = 80, rotation = 0, fontsize=20, fontweight = 'bold')
plt.legend(fontsize=14, loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **Read toothpaste sales data of each month and show it using scatterplot.**

```python
%config InlineBackend.figure_format = 'retina'
plt.figure(figsize=(14, 8))
scatter = plt.scatter(
    data['month_number'],
    data['toothpaste'],
    c=data['month_number'],
    cmap='viridis',
    s=100,
    alpha=0.9,
    edgecolor='black'
)
plt.colorbar(scatter)
plt.title('\nToothpaste Sales by Month\n', fontsize=30, fontweight='bold')
plt.xlabel('\nMonth\n', fontsize=20, fontweight='bold')
plt.ylabel('Toothpaste Units Sold', fontsize=20, fontweight='bold', labelpad = 90, rotation = 45)
plt.grid(True)
plt.show()
```

# **Read facecream and facewash product sales data and show it using barchart.**

```python
%config InlineBackend.figure_format = 'retina'
plt.figure(figsize=(12, 6))
plt.bar(data['month_number'] - 0.17, data['facecream'], width=0.4, color='lime', label='Facecream', edgecolor='black')
plt.bar(data['month_number'] + 0.1, data['facewash'], width=0.4, color='orangered', label='Facewash', edgecolor='black')
plt.title('\nFacecream and Facewash Sales by Month\n', fontsize=24, fontweight='bold')
plt.xlabel('\nMonth\n', fontsize=18, fontweight='bold')
plt.ylabel(' Units Sold', fontsize=18, fontweight='bold', labelpad = 70, rotation = 0)
plt.legend(fontsize=14)
plt.grid(True, which='both', linestyle='--', linewidth=0.7)

plt.tight_layout()
plt.show()
```

# **Read the total profit of each month and show it using the histogram to see the most common profit range.**

```python
%config InlineBackend.figure_format = 'retina'
plt.figure(figsize=(10, 6))
plt.hist(data['total_profit'], bins=10, color='cyan', edgecolor='black')
plt.title('\nDistribution of Total Profit\n', fontsize=23, fontweight='bold')
plt.xlabel('\nTotal Profit\n', fontsize=17, fontweight='bold')
plt.ylabel(' Frequency', fontsize=17, fontweight='bold', labelpad = 70, rotation = 0)
plt.grid(True, which='both', linestyle='--', linewidth=0.7)

plt.tight_layout()
plt.show()
```

# **Read all products sales data and show it using the multi-line plot.**

```python
%%R
ggplot(data, aes(x = month_number)) +
  geom_line(aes(y = facecream, color = "Facecream"), size = 1) +
  geom_line(aes(y = facewash, color = "Facewash"), size = 1) +
  geom_line(aes(y = toothpaste, color = "Toothpaste"), size = 1) +
  geom_line(aes(y = bathingsoap, color = "Bathing Soap"), size = 1) +
  geom_line(aes(y = shampoo, color = "Shampoo"), size = 1) +
  geom_line(aes(y = moisturizer, color = "Moisturizer"), size = 1) +
  labs(title = "Sales Data for All Products", x = "Month Number", y = "Sales") +
  scale_color_manual(name = "Products", values = c("Facecream" = "blue", "Facewash" = "red",
                                                   "Toothpaste" = "green", "Bathing Soap" = "orange",
                                                   "Shampoo" = "purple", "Moisturizer" = "brown"))
```

# **Read toothpaste sales data of each month and show it using scatterplot.**

# **Read facecream and facewash product sales data and show it using barchart.**

```python
%%R
data_long <- data %>%
  select(month_number, facecream, facewash) %>%
  gather(key = "Product", value = "Sales", -month_number)

ggplot(data_long, aes(x = factor(month_number), y = Sales, fill = Product)) +
  geom_bar(stat = "identity", position = "dodge") +
  labs(title = "Facecream and Facewash Sales Comparison", x = "Month Number", y = "Sales") +
  scale_fill_manual(values = c("Facecream" = "blue", "Facewash" = "red"))
```
